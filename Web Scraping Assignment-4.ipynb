```python

for i in range(1,600):
    
    names=driver.find_elements(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[1]/table/tbody/tr/td[2]/p/b/a")
    for i in names:
        try:
            name.append(i.text)
        except:
            name.append('-')
            
for i in range(1,600):
    datas=driver.find_elements(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[2]/p")
    for i in datas:
        try:
            data.append(i.text)
        except:
            data.append('-')
            
for i in range(1,600):
    tasks=driver.find_elements(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[3]/p")
    for i in tasks:
        try:
            task.append(i.text)
        except:
            task.append('-')
            
for i in range(1,600):
    attributes=driver.find_elements(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[6]/p")
    for i in attributes:
        try:
            attributetype.append(i.text)
        except:
            attributetype.append('-')

for i in range(1,600):
    no_instances=driver.find_elements(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[5]/p")
    for i in no_instances:
        try:
            no_instance.append(i.text)
        except:
            no_instance.append('-')
            
for i in range(1,600):
    no_attributes=driver.find_elements(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[6]/p")
    for i in no_attributes:
        try:
            no_attribute.append(i.text)
        except:
            no_attribute.append('-')
for i in range(1,600):
    years=driver.find_elements(By.XPATH,f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[7]/p")
    for i in years:
        try:
            year.append(i.text)
        except:
            year.append('-')
```
