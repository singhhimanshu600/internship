Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and

```python
try:
    names=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')
    for i in names:
        hostelname.append(i.text)
except NoSuchElementException:
    hostelname.append('-')
    
try:
    distance=driver.find_elements(By.XPATH,'//span[@class="description"]')
    for i in distance:
        distance_city.append(i.text)
except NoSuchElementException:
    distance_city.append('-')
    
try:
    rate=driver.find_elements(By.XPATH,'//div[@class="score orange big"]')
    for i in rate:
        rating.append(i.text)
except:
    rating.append('-')
    
try:
    total=driver.find_elements(By.XPATH,'//div[@class="reviews"]')
    for i in total:
        total_review.append(i.text)
except NoSuchElement:
    total_review.append('-')
    
try:
    overall=driver.find_elements(By.XPATH,'//div[@class="keyword"]//span')
    for i in overall:
        overall_review.append(i.text)
except NoSuchElementException:
    overall_review.append('-')
    
try:
    private=driver.find_elements(By.XPATH,'//div[@class="prices-col"]/a/div[1]')
    for i in private:
        private_price.append(i.text.replace('\n',''))
except NoSuchElementException:
    private_price.append('-')
    
try:
    dorms=driver.find_elements(By.XPATH,'//div[@class="prices-col"]/a/div[2]')
    for i in dorms:
        dorms_price.append(i.text.replace('\n',''))
except NoSuchElementException:
    dorms_price.append('-')
    
try:
    facilities=driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
    for i in facilities:
        facility.append(i.text)
except NoSuchElementException:
    facility.append('-')
```
